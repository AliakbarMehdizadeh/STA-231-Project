```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(x, r):
    return r * x * (1 - x)

def generate_bifurcation_diagram(r_values, iterations=10000, transient=500):
    x = 0.5  # Initial condition

    # Iterate through different values of r
    bifurcation_data = []
    for r in r_values:
        for _ in range(transient):
            x = logistic_map(x, r)

        for _ in range(iterations):
            x = logistic_map(x, r)
            bifurcation_data.append([r, x])

    return np.array(bifurcation_data)

# Define the range of r values
r_values = np.linspace(2.5, 4.0, 1000)

# Generate bifurcation diagram data
bifurcation_data = generate_bifurcation_diagram(r_values)
```

```python
# Plotting
plt.figure(figsize=(10, 6),dpi=100)
plt.scatter(bifurcation_data[:, 0], bifurcation_data[:, 1], s=0.001, c='blue', marker='.')
plt.xlabel('r (growth rate)')
plt.ylabel('Population')
plt.title('Logistic Map Bifurcation Diagram')
plt.show()
```
