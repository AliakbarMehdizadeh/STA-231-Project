```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(x, r):
    return r * x * (1 - x)

def logistic_map_derivative(x, r):
    return r * (1 - 2 * x)

def lyapunov_exponent(x, r, iterations=1000):
    sum_log_derivative = 0
    for _ in range(iterations):
        derivative = logistic_map_derivative(x, r)
        sum_log_derivative += np.log(np.abs(derivative) + 1e-10)
        x = logistic_map(x, r)
    return sum_log_derivative / iterations

# Generate a range of r values
r_values = np.linspace(3, 4, 1000)

# Calculate Lyapunov exponent for each r
lyapunov_exponents = [lyapunov_exponent(0.5, r) for r in r_values]
```

```python
# Plotting
plt.figure(figsize=(10, 6), dpi = 100)
plt.plot(r_values, lyapunov_exponents, label='Lyapunov Exponent', color='blue')
plt.axhline(0, color='black', linestyle='--', linewidth=0.5, label='Zero Line')

plt.title('Lyapunov Exponent for Logistic Map')
plt.xlabel('r (growth rate)')
plt.ylabel('Lyapunov Exponent')
plt.legend()
plt.show()
```
